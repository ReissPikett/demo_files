### Rolling averages and 'long' holding
Here we will use the rolling averages feature of pandas dataframes and make a simple 'long' stock holding strategy. A Rolling average of x period is the average of the following x points of any given point, for each data point in the data set. We will use two rolling averages, and assume that if the one with the smaller averaging period is higher that that with a greater averaging period, then we should hold 1 stock while this is true.

## Log Returns Distribution and Predicting Percentage Increase
Here we will show the log returns of the proportion of the Adjusted CLose price each day to the previous day, show the distribution of retults, and make some predictions. First we will divide each days closing price by that of the day before, and take the log of each result. We use the log values for multiple reasons, one being that here we assume that stock price data has a log-normal distribution, and is therefore easier to make calculations and inferences if we use the log of the results.  

```python
ms['LogReturn'] = np.log(ms['Adj. Close']/ms['Adj. Close'].shift(1))
ms.describe()
```

```python
# Plot a histogram to show the distribution of log return of Microsoft's stock. 
# Using the Scipy.stats package, we can plot a normal distribution of the sample mean and standard deviation of the data

mu = ms['LogReturn'].mean()
sigma = ms['LogReturn'].std(ddof=1)

density = pd.DataFrame()
density['x'] = np.arange(ms['LogReturn'].min()-0.01, ms['LogReturn'].max()+0.01, 0.001)
density['pdf'] = norm.pdf(density['x'], mu, sigma)

ms['LogReturn'].hist(bins=200, figsize=(15, 8), density = True)
plt.plot(density['x'], density['pdf'], color = 'red')
plt.show()
```

We can see here that the data is close to a normal distribution, though has slightly higher frequency at the tails.

```python
# Probability that the stock price of microsoft will drop over 5% in a day. Using scipy package and our normal distribution
# to make predictions 
probability_return1 = norm.cdf(-0.05, mu, sigma)
print('The probability that the stock will drop more than 5% in one day is: '+ str(100*np.round(probability_return1,2))+' %')
```

Now we will do the same for return gains, over longer period of time. This will be useful for predicting the future returns of purchasing a stock. The prediction of the sample mean over a longer period is obtained by multiplying the sampled mean by the period, and the period standard deviation by multiplying the sample standard deviation by the square root of the period. As there are aproximately 252 working days in the year, we consider the number of days in the period in question as the proportion of the year (e.g. 0.5 years) multiplied by 252 working days.

Here we look at the percentage that holding a certain stock will return a given percentage increase/loss over a given period of time:

```python
def probability_returns(percentage_change, proportion_of_year, data): 
    """
    Function for returning the probability that the price of a certain stock will have a given
    increase/decrease in percentage over a given period of time:
    
    Arguments
    percentage_change = percentage change of stock value over period
    proportion_of_year = fraction of year. e.g. for 3 months proportion of year = 1/4
    data - data set
    
    Returns
    probability - probability that the stock 'data' will change by 'percentage_increase' over 'proportion_of_year*252' days
    
    """
    period=proportion_of_year*252
    log_ret=np.log(data['Adj. Close']/data['Adj. Close'].shift(1))
    mu=log_ret.mean()*period
    sigma=log_ret.std(ddof=1)*(period**0.5)
    
    if percentage_change>=0:
        probability_return = 1-norm.cdf(percentage_change, mu, sigma)
        print('The probability of the stock price increasing by more than', np.abs(np.round(percentage_change,3))*100,'% over', (12*proportion_of_year),'months is ', 100*np.round(probability_return,2),' %')

    elif percentage_change<0:
        probability_return = norm.cdf(percentage_change, mu, sigma)
        print('The probability of the stock price decreasing by more than', np.abs(np.round(percentage_change,3)*100),'% over', (12*proportion_of_year),'months is ', 100*np.round(probability_return,2),' %')
    return probability_return
```

```python
percentage_change, proportion_of_year = -20/100,2/12
predicted_probability = probability_returns(percentage_change,proportion_of_year,ms);
```

Here we can look at the data resampled over 15 Days to see if this formula holds true by comparing our predicted percentage change probability for that period and the actual resampled percentage change probability. To do this we have to make an assumption that daily returns are independant, which is generally incorrect, but for now is acceptable for our basic model and short time period.

```python
percentage_change, proportion_of_year, gain =5/100,1/24,1
predicted_probability = probability_returns(percentage_change,proportion_of_year,ms);
```

```python
ms2=ms.drop('LogReturn',axis=1).resample('15D').mean()
ms2['Month logret']=np.log(ms2['Adj. Close']/ms2['Adj. Close'].shift(1))
mean2=ms2['Month logret'].mean()
std2=ms2['Month logret'].std(ddof=1)
resampled_probability=1-norm.cdf(percentage_change,mean2,std2)

print('Compare statistics:')
print('Predicted Probability of', percentage_change*100 ,'% change:', np.round(predicted_probability,3)*100, '%')
print('Resampled Probability of', percentage_change*100 ,'% change:', np.round(resampled_probability,3)*100, '%')
print('Predicted mean:', np.round(mean2,3), 'Predicted std:' ,np.round(std2,3))
print('Resampled mean:', np.round(mu*proportion_of_year*252,3), 'Resampled period Standard deviation', np.round(sigma*proportion_of_year*252**0.5,3))
```

As we can see, there is some similarity in the predictions of longer periods with respect to the resampling of periods, but the normal distribution will need to be improved, perhaps with a 'fat-tailed' distrubution.

# Value at Risk (VaR)
This is to determine that, continuing with our a normal distribution, though with certain adjustments this can be used for any model, what is the maximum stock price gain or loss for a given percentage likelihood. For example, what is the minimum value of stock price decrease that has a probability of less than or equal to 5% of occurring

```python
risk = 0.05
VaR = norm.ppf(risk, mu, sigma)
print('Here we can see that the Value at Risk for a given day at '+str(risk*100)+ '% risk level in a given day is',np.round(VaR,3)*100,'% change in stock price')
```

Again, we can also do this for predictions over a period:

```python
VaR_month = norm.ppf(0.05, mean2, std2)
print('The Value at Risk at', risk*100, '% risk level over a period of 15 days is', np.round(VaR_month,3)*100, '% change in stock price')
```

```python
# VaR at various Quantiles 
# 5% quantile
print(f'5% quantile: {np.round(norm.ppf(0.05,mu,sigma),3)*100}%')
# 25% quantile
print(f'25% quantile: {np.round(norm.ppf(0.25,mu,sigma),3)*100}%')
# 50% quantile
print(f'50% quantile: {np.round(norm.ppf(0.50,mu,sigma),3)*100}%')
# 75% quantile
print(f'75% quantile: {np.round(norm.ppf(0.75,mu,sigma),3)*100}%')
# 95% quantile
print(f'95% quantile: {np.round(norm.ppf(0.95,mu,sigma),3)*100}%')
```

Here we can see the VaR for various quantiles of the data. The VaR for the positive percentages show the maximum percentage price increase that is likely to occur at that percentage risk. The corrollary of this is that if the VaR at 95% is 4.9%, there is a 5% chance that the price will increase more than 4.9%.

# 2) Sampling and Hypothesis Testing
## Estimate the average stock return with 90% Confidence Interval

A Confidence Interval of x% tells us that there will be x% chance that the average stock return lies in a certain range. Let us build a 90% Confidence Interval for Microsoft stock

```python
# 90% confidence interval for log return of microsoft stock
sample_size = ms['LogReturn'].shape[0]
sample_mean = ms['LogReturn'].mean()
sample_std = ms['LogReturn'].std(ddof=1) / sample_size**0.5

# left and right quantile
z_left = norm.ppf(0.05)
z_right = norm.ppf(0.95)

# upper and lower bound
interval_left = z_left*sample_std+sample_mean
interval_right = z_right*sample_std+sample_mean
```

```python
# 90% confidence interval tells you that there will be 90% chance that the average stock return lies between "interval_left"
# and "interval_right".

print('90% confidence interval is [',np.round(interval_left*100, 3),'%,', np.round(interval_right*100,3),'%]')
```

This tells us that there is a 90% chance that the price change over one day will be between -0.02% and 0.10%

## Steps involved in testing a claim by hypothesis testing
### Set hypothesis
$H_0 : \mu = 0$

$H_a : \mu \neq 0$

H_0 is the hypothesis that the average stock return is zero. 
H_a is the hypothesis that the average stock return is not zero. 

We can test the data by taking the mean, standard deviation and sample size of the log data, and determining whether the mean deviates from the null hypothesis with a certain confidence. We can do this at an arbitrary confidence level, though 10% is a sensible value. We calculate the number of standard deviations that the mean is from zero, also known as the z-statistic, and determine the probability that this mean occured by chance in a normal distribution of the given standard deviation about zero. If this value is outside of our confident level, then we reject the Null Hypothesis. 

```python
confidence = 0.10
null_hypothesis_value = 0
```

```python
# Sample statistics
sample_mean = ms['LogReturn'].mean()
sample_size = ms['LogReturn'].shape[0]
sample_std = ms['LogReturn'].std(ddof=1)/sample_size**0.5
zhat = (sample_mean-null_hypothesis_value)/(sample_std) #Z statistic- number of standard deviations from the mean

#Confidence interval

zleft = norm.ppf(confidence/2,0,1)
zright = -zleft

print('z-statistic of sample mean: ', np.round(zhat,3))
print('Interval over which we accept the Null Hypothesis:')
print('[',np.round(zleft,3),',',np.round(zright,3),']')
reject=False
response={True:'reject', False:'not reject'}
if zhat > zright or zhat <zleft:
    reject=True
print('At the confidence level of', confidence*100, '%, we shall '+response[reject]+' the Null Hypothesis')
```

## One tail test for potential price gain on prospective stock

$H_0 : \mu \leq 0$ 
$H_a : \mu > 0$

```python
alpha = 0.1
zright = norm.ppf(1-alpha, 0, 1)
reject=False
response={True:'reject', False:'not reject'}
if zhat > zright or zhat <zleft:
    reject=True
print("Right tail confidence boundary: ", np.round(zright,3))
print('At the confidence level of', alpha, ', we shall '+response[reject]+' the Null Hypothesis')
```

### P-value
The statistical inverse of the above method, we calculate the p-score, and determine the probability that the sample mean is due to randomness or if it statistically significantly greater than our hypothesis boundary value (zero), by asking if the p-value it is greater than our confidence value

```python
p = 1-norm.cdf(zhat, 0, 1)
print('Right boundary: ',np.round(p,4))
print('At confidence level of {}, Reject: {}'.format(confidence, p < confidence))
```

At this confidence level, we accept the null hypothesis and may consider this stock as not worthy of investment

# 3) Simple linear regression and Model Diagnostics
$y_i = \beta_0 + \beta_1 * x_i + \epsilon_i $
### We shall base on the association between LSTAT and MEDV and create a simple linear regression model. Let's use python in estimating the values of B0 and B1 (intercept and slope)

Intercept and slope trial, demonstrating pandas plot functions, and calculating the error of our guess. We can calculate the error using the Root Mean Squared Error(RMSE). RMSE is the square root of the sum of squared errors for each point in the set.
We call our guess b0, b1, and assign the value of b0, b1 to get a straight line that can describe our data

```python
b0 = -45.0
b1 = 11
housing['GuessResponse'] = b0 + b1*housing['RM']

# Guess error
housing['observederror'] = housing['MEDV'] - housing['GuessResponse']
RMSE = np.sqrt((housing['observederror']**2).sum())

# plottting the estimated line together with the points
plt.figure(figsize=(12, 9))
plt.title('')
plt.scatter(housing['RM'], housing['MEDV'], color='g', label='Observed')
plt.plot(housing['RM'], housing['GuessResponse'], color='red', label='GuessResponse')
plt.legend()
plt.xlim(housing['RM'].min()-2, housing['RM'].max()+2)
plt.ylim(housing['MEDV'].min()-2, housing['MEDV'].max()+2)
plt.show()
print('RMSE error is {}'.format(np.round(RMSE,1))) 
```

Let us see if we can improve our guess with regression tools

# Least squares estimates
Finding the Mean Squared Error (MSE) value, by caluculating the sum of the squared differences between the observed values and the predicted values, and using the statsmodels Ordinary Least Squares (OLS) method to minimise this sum.

```python
# Import statsmodels ols method and set variables
import statsmodels.formula.api as smf
import warnings
warnings.filterwarnings("ignore")

formula = 'MEDV~RM' # dependant and independent variables
model = smf.ols(formula=formula, data=housing).fit() # calling statsmodels OLS function and fitting to data

# Here are the calculated intercept and slope by the OLS estimation
b0_ols = model.params[0] # calculated intercept 
b1_ols = model.params[1] # calculated gradient

#Formula of our linear model and its error
housing['BestResponse'] = b0_ols + b1_ols*housing['RM']
housing['Squared error'] = (housing['MEDV'] - housing['BestResponse'])**2
RMSE_OLS = np.sqrt(housing['Squared error'].sum())

# plotting estimated line together with the points
plt.figure(figsize=(12, 9))
plt.title('Sum of sqaured error is {}'.format((housing['Squared error']).sum()))
plt.scatter(housing['RM'], housing['MEDV'], color='g', label='Observed')
plt.plot(housing['RM'], housing['GuessResponse'], color='red', label='GuessResponse')
plt.plot(housing['RM'], housing['BestResponse'], color='yellow', label='BestResponse')
plt.legend()
plt.xlim(housing['RM'].min()-2, housing['RM'].max()+2)
plt.ylim(housing['MEDV'].min()-2, housing['MEDV'].max()+2)
plt.show()
```

# Summary table
We can refer to some specific values of the summary table for our model to evaluate the performance, such as the P-value, which tells us the confidence at which we can say the relationship is not due to chance, and the R-squared value, which represents the proportion of the variance for the dependent variable that's explained by the independent variable. Here we have a middle to low R-value, which suggests that our model is rather good at predicting the general trend but does not capture enough of the variance of MDEV to make exact predictions. Perhaps considering more variables will provide more accurate predictions

# Model Diagnostics
Let us now try to determine a relationship between LSTAT and MEDV in our housing data. There are various assumptions that are made in order to decide that a linear regression model is suitable for representing the relationship between variables. We have Linearity, Independance, Normality and Equal Variance. We will now check the variable data and see if they conform to our assumptions, therefore showing if linear regression is a valid choice of model

## Linearity
Linearity is the behaviour that the response variable is equal to a first order polynomial equation of the predictor variable, generalised in our statsmodel equation above, plus some error term.

```python
# We can check the scatter plot to have a fast check
housing.plot(kind='scatter', x='LSTAT', y='MEDV', figsize=(10, 7), color='g')
```

The data seems to have more of a higher order polynomial trend, or perhaps is better represented by a log function. The data therefore does not pass our linearity test.

## Independance
Error independance is that there is no serial correlation between subsequent errors. Two ways of determining this is by showing a residual error plot between our response prediction and the data, and seeing if there is a visible relationship. Another is by using the Durbin-Watson statistic included in the statsmodels metrics

```python
# Get all errors (residuals)
housing['error'] = housing['MEDV'] - housing['BestResponse']

# Method 1: Residual vs order plot
# error vs order plot (Residual vs order) as a fast check 
plt.figure(figsize=(12, 7))
plt.title('Residual vs order')
plt.plot(housing.index, housing['error'], color='purple')
plt.axhline(y=0, color='red')
plt.show()
```

It is difficult to determine the independance of this data for certain visually, it is possible that the serial errors have an almost sinusoidal relationship, but this may be random. We must therefore use the Durbin-Watson statistic. 
The rule of thumb: test statistic values in the range of 1.5 to 2.5 are relatively normal, <1.5 implies a positive correlation between subsequent errors, and >2.5 a negative correlation

```python
# Method 2: Durbin Watson Test
model.summary()
```

As we can see the Durbin-Watson statistic is significantly lower than our 1.5 limit, and so serial errors are positively correlated. The data therefore does not conform to Independance

## Normality
This is the assumption that the errors of the model has a normal distribution. we can show this with a histogram or a normal Q-Q plot, where the theoretical quantiles, representing typical values over a normal distribution with a mean and standard deviation of the sample mean and standard deviation, and the Ordered Values which are the actual values corresponding to that quantile. These should have a linear, one-to-one relationship for the data to be considered Normal.  

```python
import scipy.stats as stats
z = (housing['error'] - housing['error'].mean())/housing['error'].std(ddof=1)
z.hist(bins=14)
plt.show()
stats.probplot(z, dist='norm', plot=plt)
plt.title('Normal Q-Q plot')
plt.show()
```

As we can see, the data follows a somewhat normal distribution about the interquartile interval, but deviates at the extreme values, particularly the positive values, and so has a postive skew.

## Equal Variance
Equal variance is the requirement that the variance of the data is roughly the same throughout, and does not have some correlation with the predictor variable.

```python
# Residual vs predictor plot
housing.plot(kind='scatter', x='LSTAT', y='error', figsize=(12, 7), color='green')
plt.title('Residual vs predictor')
plt.axhline(y=0, color='red')
plt.show()
```

As we can see, there is much variation in the variance throughout this data, with, for example, smaller values at LSTAT=8 and larger for LSTAT > 25.

### As the regression model (MEDV~LSTAT) violates all four assumptions, we cannot make statistical inference using this model, and know that it requires further development.

```python
# RMSE - Root Mean Squared Error, Adjusted R^2. This is required for our multiple predictor variables
def adjustedMetric(data, model, model_k, yname):
    data['yhat'] = model.predict(data)
    SST = ((data[yname] - data[yname].mean())**2).sum()
    SSR = ((data['yhat'] - data[yname].mean())**2).sum()
    SSE = ((data[yname] - data['yhat'])**2).sum()
    r2 = SSR/SST
    adjustR2 = 1 - (1-r2)*(data.shape[0] - 1)/(data.shape[0] -model_k -1)
    RMSE = (SSE/(data.shape[0] -model_k -1))**0.5
    return adjustR2, RMSE
```

```python
def assessTable(test, train, model, model_k, yname):
    r2test, RMSEtest = adjustedMetric(test, model, model_k, yname)
    r2train, RMSEtrain = adjustedMetric(train, model, model_k, yname)
    assessment = pd.DataFrame(index=['R2', 'RMSE'], columns=['Train', 'Test'])
    assessment['Train'] = [r2train, RMSEtrain]
    assessment['Test'] = [r2test, RMSEtest]
    return assessment
```

```python
# Get the assement table for our model
assessTable(Test, Train, lm, 9, 'spy')
```

## Conclusion

The results suggest that there is some predictive power in our multivariable model, with $7$% of the data covariance being explained by our predictions, which, for stock prices, is not bad at all. It is worth noting that, as the Root Mean Squared Error (RMSE) is greater for our Test sample than it is for our Train sample, there is some overfitting in our model, which is often the case which when considering such 'noisy' data as stock prices.

```python
Train['Wealth'] = Train['Wealth'] + Train.loc[Train.index[0], 'Price']
Test['Wealth'] = Test['Wealth'] + Test.loc[Test.index[0], 'Price']
```

```python
# Sharpe Ratio on Train data
Train['Return'] = np.log(Train['Wealth']) - np.log(Train['Wealth'].shift(1))
dailyr = Train['Return'].dropna()

print('Daily Sharpe Ratio is ', dailyr.mean()/dailyr.std(ddof=1))
print('Yearly Sharpe Ratio is ', (252**0.5)*dailyr.mean()/dailyr.std(ddof=1))
```

```python
# Sharpe Ratio in Test data
Test['Return'] = np.log(Test['Wealth']/Test['Wealth'].shift(1))
dailyr = Test['Return'].dropna()
```

```python
print('Daily Sharpe Ratio is ', dailyr.mean()/dailyr.std(ddof=1))
print('Yearly Sharpe Ratio is ', (252**0.5)*dailyr.mean()/dailyr.std(ddof=1))
```

```python
# Maximum Drawdown in Train data
Train['Peak'] = Train['Wealth'].cummax()
Train['Drawdown'] = (Train['Peak'] - Train['Wealth'])/Train['Peak']
print('Maximum Drawdown in Train is ', Train['Drawdown'].max())
```

```python
# Maximum Drawdown in Test data
Test['Peak'] = Test['Wealth'].cummax()
Test['Drawdown'] = (Test['Peak'] - Test['Wealth'])/Test['Peak']
print('Maximum Drawdown in Test is ', Test['Drawdown'].max())
```
